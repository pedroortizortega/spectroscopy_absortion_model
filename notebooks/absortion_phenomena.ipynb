# Spectroscopy Absortion Description

Los fenómenos de espectroscopía son ampliamente conocidos por la interacción entra la luz y materia. Para ello es necesario que un haz de luz incidente impacte en alguna muestra y está tenga alguno de los posibles, y generales, fenómenos:

* Absorción
* Emisión
* Transmisión

En lo que respecta a espectroscopía, y en particular al trabajo presente, nos enfocaremos en el fenómeno de absorción. Es decir, como el material objetivo absorbe un porcentaje de la luz incidente de una fuenta dada y con ello obtener información de la muestra debido a la cantidad de energía absorbida. 

## 1.- Modelo simplificado - AM1

Existen varias formas de aproximar teoricamente lo modelos de absorción, pero empezaremos haciendo las siguientes supocisiones:
* El haz de luz que impacta sobre la muestra se pueda modelar como un dipolo electrico polarizado
* El hamiltoneano base será el de Hidrógeno molecular (H2)
* Se aplicará método de perturbaciones a primer y segundo término
* Asumiremos la hipótesis de Somerfeld - Oppenheimer, es decir, suponemos núcleos fijos respecto al marco inercial de los electrones
